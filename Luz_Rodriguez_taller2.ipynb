```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
```

```python
G = nx.read_gml('dolphins.gml')
```

```python
G
```

```python
print("Número de nodos:", G.number_of_nodes())
print("Número de redes:", G.number_of_edges())
```

```python
list(G.nodes)
```

```python
list(G.edges)
```

```python
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(G, with_labels=True,
        pos=nx.fruchterman_reingold_layout(G, seed=12),
        node_size=60, 
        node_color="purple", 
        node_shape="h", 
        alpha=0.5, 
        linewidths=50,
        font_weight='bold')
plt.show()
```

```python
nt = Network('1000px', '1000px', notebook = True)
nt.from_nx(G)
nt.show('nx.html')
```

# Framework de visulaización

## Datos

Red social no dirigida de asociaciones frecuentes entre 62 delfines en una comunidad que vive en Doubtful Sound, Nueva Zelanda, los enlaces entre nodos (Delfines) representan las relaciones mas dinamicas y fluidas. 

Referencias
D. Lusseau, K. Schneider, OJ Boisseau, P. Haase, E. Slooten y SM Dawson, Behavioral Ecology and Sociobiology 54 , 396-405 (2003). 

## **¿Que?**

El gráfico de red muestra una población de 62 delfines, estos representados en los nodos, los cuales se conectan mediante los enlaces que reprentan, relaciones dinamicas y fluidas entre los delfines. Se puede observar que en general el compañerismo en el grupo de delfines es alta, pues no mas de 15 delfines tienen 2 o menos relaciones con otros delfines, ademas tambien se observa que los delfines Topless, Grin, Scabs, Sn4, kringel y web son los que mas relaciones presentan. Por otro lado se entiende que el dataset es de tipo Red, ya que cuenta con 62 nodos y 149 enlaces, por ultimo los datos con los que se cuenta son estaticos.

* Tipo de Dataset: Arboles y Redes

* Tipos de datos: 

 - Nodos 
 
 - Enlaces

## **¿Por que?**

En este caso los datos ya vienen divididos por nodos y enlaces no dirigidos, por lo tanto no es necesaria una transformación de datos, sin embargo hay que tener cuenta que para obentener estos datos, otra base con la información original y con mas variables tuvo que ser tranformada de modo que se mostraran los nodos y enlaces, y que ademas estos fueran los datos principales a utilizar.

Por otro lado, este diagrama de enlace de nodo tiene como acción o proposito principal **presentar**, ya que lo que nos muestra principalmente son las relaciones dinamicas y fluidas entre los delfines, tambien el diagrama **localiza** e **identifica** ya que se puede observar que Topless, Grin, Scabs, Sn4, kringel y web son los delfines que muestran mas vinculos dinamicos con otros delfines, y que lo nodos (delfines) que estan en las puntas suelen ser aquellos que presentan pocos niculos con otros delfines.

Asi mismo el objetivo de este diagrama es encontrar que delfines tienen mas realciones entre ellos y encontrar delfines de un grupo que son puente de interacción con otro grupo de delfines.

**1. Actions**

* Analizar: 

 - Consumir: Presentar

* Especificas:

 - Buscar: Localizar
 - Consulta: Identificar


**2. Objetivos**

* Datos de Red: 

 - Relaciones entre nodos
 - Puentes

## **¿Como?**

Para este diagrama, la base no necesito de codificación ya que esta ya venia con los nodos y enlaces identificados, por otro lado el diagrama es manipulable, pues se puede **navegar** en el y **Seleccionar** nodos y enlaces, pues se opto por hacerlo interactivo, ya que se tienen varios nodos y muchos mas enlaces y de esta forma es mas facil identificar con que delfines se conecta un delfin en especifico.

## Marcas y Canales

Marca:
Las marcas para este diagrama son los objetos de puntos y lineas, los primeros representan los nodos que para este caso son los delfines y los segundos son los enlaces que concetan estos nodos para representar las interacciones dinamicas entre estos.

Canal:
Posición espacial: sin significado codificado directamente.

## Cumplimiento de reglas generales

Para este caso no fueron tomadas en cuenta las reglas relacionadas con el 3D, ya que el diagrama está en 2D y está justificado debido a que la base cuenta con bastantes nodos y enlaces y de esta forma se puede visualizar mayor cantidad de esta información en un solo espacio y es adecuado para ver los enlaces más directamente.

La regla de los ojos le gana a la memoria también la cumple este diagrama, pues este se genera en una sola vista y los usuarios pueden fácilmente visualizar las conexiones de un nodo en específico sin dejar de lado los no seleccionados. la regla Resolución sobre la inmersión no se toma en cuenta y finalmente la regla primero overview, zoom y filtro se cumple ya que la visualización da una en un primer momento una visión general de la topología de la red, como se conectan los nodos, pero también podemos si es necesario ver más de cerca estas conexione lo podemos hacer gracias al zoom y la interacción que presenta este diagrama


## Recomendaciones

Una de las mejoras que pienso que podria dar mas informacióna al visualizar el diagrama, seria que los nodos por color diferenciaran machos y hembras de delfines, ya que esto podria decirnos porque algunos delfines tiene mas conexiones que otras, por ejemplo algunos estudios indican que las hembras suelen entablar mas conexiones para asi tener mas protección de sus crias, seria interesante poder representar esto en el diagrama.
